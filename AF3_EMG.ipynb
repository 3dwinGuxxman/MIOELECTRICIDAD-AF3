**AF3. EMG**


```python
# Simulacin de seal EMG del bceps braquial
import numpy as np
import matplotlib.pyplot as plt

# Parmetros de la seal EMG
fs = 2000  # Frecuencia de muestreo en Hz
t = np.linspace(0, 1, fs)  # 1 segundo de duracin
# Simulacin de seal EMG: ruido blanco filtrado + picos aleatorios
np.random.seed(42)
emg_signal = np.random.normal(0, 0.5, fs)
emg_signal += np.sin(2 * np.pi * 50 * t) * 0.1  # componente muscular
emg_signal *= np.exp(-t * 2)  # decaimiento para simular fatiga

# Grfica
plt.figure(figsize=(12, 4))
plt.plot(t, emg_signal, color='black')
plt.title('Electromiograma simulado del msculo bceps braquial')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (mV)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
