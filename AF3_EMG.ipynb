**AF3. EMG**


**BICEPS BRAQUIAL**

```python
# Simulación de señal EMG del bíceps braquial
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal EMG
fs = 2000  # Frecuencia de muestreo en Hz
t = np.linspace(0, 1, fs)  # 1 segundo de duración
# Simulación de señal EMG: ruido blanco filtrado + picos aleatorios
np.random.seed(42)
emg_signal = np.random.normal(0, 0.5, fs)
emg_signal += np.sin(2 * np.pi * 50 * t) * 0.1  # componente muscular
emg_signal *= np.exp(-t * 2)  # decaimiento para simular fatiga

# Gráfica
plt.figure(figsize=(12, 4))
plt.plot(t, emg_signal, color='black')
plt.title('Electromiograma simulado del músculo bíceps braquial')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (mV)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**TIBIAL ANTERIOR**

```python
# Simulación de señal EMG del músculo tibial anterior
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal EMG
fs = 2000  # Frecuencia de muestreo en Hz
t = np.linspace(0, 1, fs)  # 1 segundo de duración

# Simulación de señal EMG: contracción intermitente con ruido
np.random.seed(123)
emg_tibial = np.random.normal(0, 0.4, fs)
emg_tibial += np.sin(2 * np.pi * 30 * t) * 0.2  # componente muscular
emg_tibial *= (np.sin(2 * np.pi * 2 * t) > 0).astype(float)  # contracción intermitente

# Gráfica
plt.figure(figsize=(12, 4))
plt.plot(t, emg_tibial, color='darkblue')
plt.title('Electromiograma simulado del músculo tibial anterior')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (mV)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
